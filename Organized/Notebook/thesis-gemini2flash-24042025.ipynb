```python
def get_response_openrouter(prompt, model="google/gemini-2.0-flash-001"):
    try:
        completion = client.chat.completions.create(
            extra_headers={
                "HTTP-Referer": "<YOUR_SITE_URL>",  # Optional for tracking
                "X-Title": "<YOUR_SITE_NAME>",      # Optional for rankings
            },
            model=model,
            messages=[
                {"role": "user", "content": prompt}
            ]
        )
        return completion.choices[0].message.content
    except Exception as e:
        print("Error using OpenRouter:", e)
        return None
```

```python
def generate_user_stories_zero_shot(input_file, output_file):
    # Read the CSV
    df = pd.read_csv(input_file)

    if 'data' not in df.columns:
        print("Error: 'data' column not found in the input file.")
        return

    user_stories = []
    for index, row in df.iterrows():
        user_input = row['data']
        print(f"Generating user story for review {index + 1}...")

        prompt = f"Convert the following user review into a user story from the user's perspective: {user_input}"
        user_story = get_response_openrouter(prompt)

        user_stories.append(user_story if user_story else "Error generating user story.")

    df['user_story'] = user_stories
    df.to_csv(output_file, index=False)
    print(f"User stories generated and saved to {output_file}")
```

```python
def generate_user_stories_one_shot(input_file, output_file):
    # Example one-shot pair
    example_review = "I wish the app allowed dark mode and had fewer steps to book a ticket. It's hard to navigate right now."
    example_user_story = "As a user, I want a dark mode option and a simpler ticket booking flow so that I can use the app comfortably and efficiently."

    # Read the CSV
    df = pd.read_csv(input_file)

    if 'data' not in df.columns:
        print("Error: 'data' column not found in the input file.")
        return

    user_stories = []
    for index, row in df.iterrows():
        user_input = row['data']
        print(f"Generating one-shot user story for review {index + 1}...")

        prompt = f"""Here is an example of converting a user review into a user story:
        
Review: "{example_review}"  
User Story: "{example_user_story}"

Now convert this review into a user story:

Review: "{user_input}"  
User Story:"""

        user_story = get_response_openrouter(prompt)
        user_stories.append(user_story if user_story else "Error generating user story.")

    df['user_story'] = user_stories
    df.to_csv(output_file, index=False)
    print(f"One-shot user stories generated and saved to {output_file}")
```

```python
# Example usage — modify paths as needed
input_file = "/kaggle/input/filtered-dataset-mini-bar/filtered.csv"
output_file = "/kaggle/working/gemini2_flash_user_stories_one_shot.csv"

generate_user_stories_one_shot(input_file, output_file)
```

```python
def generate_user_stories_two_shot(input_file, output_file):
    # Two example review-user story pairs
    example_review_1 = "I wish the app allowed dark mode and had fewer steps to book a ticket. It's hard to navigate right now."
    example_user_story_1 = "As a user, I want a dark mode option and a simpler ticket booking flow so that I can use the app comfortably and efficiently."

    example_review_2 = "The notifications are too frequent and there's no way to turn them off. It's really annoying."
    example_user_story_2 = "As a user, I want the option to manage or mute notifications so that I can control how often I'm interrupted."

    # Read input CSV
    df = pd.read_csv(input_file)

    if 'data' not in df.columns:
        print("Error: 'data' column not found in the input file.")
        return

    user_stories = []
    for index, row in df.iterrows():
        user_input = row['data']
        print(f"Generating two-shot user story for review {index + 1}...")

        # Build two-shot prompt
        prompt = f"""Here are examples of converting user reviews into user stories:
        
Review: "{example_review_1}"  
User Story: "{example_user_story_1}"

Review: "{example_review_2}"  
User Story: "{example_user_story_2}"

Now convert this review into a user story:

Review: "{user_input}"  
User Story:"""

        user_story = get_response_openrouter(prompt)
        user_stories.append(user_story if user_story else "Error generating user story.")

    df['user_story'] = user_stories
    df.to_csv(output_file, index=False)
    print(f"Two-shot user stories generated and saved to {output_file}")
```

```python
# Example usage — modify paths as needed
input_file = "/kaggle/input/filtered-dataset-mini-bar/filtered.csv"
output_file = "/kaggle/working/gemini2_flash_user_stories_two_shot.csv"

generate_user_stories_two_shot(input_file, output_file)
```
